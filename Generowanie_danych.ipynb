```python
t_range = np.arange(1,2,0.2)
t_range = np.append(t_range,np.arange(2,2.6,0.1))
t_range = np.append(t_range,np.arange(2.6,3.5,0.2))
```

```python
m10_zakres = np.load(r'C:\\Users\\mazur\\OneDrive\\Dokumenty\\GitHub\\Fizyka-ukladow-zlozonych\\danem\\m10_zakres.npy', allow_pickle=True)
m10=[]
for i in m10_zakres:
    m10.append(sum(abs(k) for k in i[10**3:10**4])/len(i[10**3:10**4]))
```

```python
m50_zakres = np.load(r'C:\\Users\\mazur\\OneDrive\\Dokumenty\\GitHub\\Fizyka-ukladow-zlozonych\\danem\\m50_zakres.npy', allow_pickle=True)
m50=[]
for i in m50_zakres:
    m50.append(sum(abs(k) for k in i[10**3:10**4])/len(i[10**3:10**4]))
```

```python
m100_zakres = np.load(r'C:\\Users\\mazur\\OneDrive\\Dokumenty\\GitHub\\Fizyka-ukladow-zlozonych\\danem\\m100_zakres.npy', allow_pickle=True)
m100=[]
for i in m100_zakres:
    m100.append(sum(abs(k) for k in i[10**3:10**4])/len(i[10**3:10**4]))
```

```python
plt.plot(t_range,m10, '-o', color='hotpink', label='L=10')
plt.plot(t_range,m50, '-o', color='purple', label='L=50')
plt.plot(t_range,m100, '-o', color='skyblue', label='L=100')
plt.grid()
plt.legend()
plt.show()
```

```python
m10_2=[]
for i in m10_zakres:
    m10_2.append(sum(k**2 for k in i[10**3:10**4])/len(i[10**3:10**4]))
pod_10 = []
for i in range(len(m10_2)):
    pod_10.append((m10_2[i]-(m10[i])**2)*10**2/t_range[i])
```

```python
m50_2=[]
for i in m50_zakres:
    m50_2.append(sum(k**2 for k in i[10**3:10**4])/len(i[10**3:10**4]))
pod_50 = []
for i in range(len(m50_2)):
    pod_50.append((m50_2[i]-(m50[i])**2)*50**2/t_range[i])
```

```python
m100_2=[]
for i in m100_zakres:
    m100_2.append(sum(k**2 for k in i[10**3:10**4])/len(i[10**3:10**4]))
pod_100 = []
for i in range(len(m100_2)):
    pod_100.append((m100_2[i]-(m100[i])**2)*100**2/t_range[i])
```

```python
plt.plot(t_range,pod_10, '-o', color='hotpink', label='L=10')
plt.plot(t_range,pod_50, '-o', color='purple', label='L=50')
plt.plot(t_range,pod_100, '-o', color='skyblue', label='L=100')
plt.grid()
plt.legend()
plt.show()
```
