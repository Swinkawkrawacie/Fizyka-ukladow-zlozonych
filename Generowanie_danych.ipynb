```python
t_range = np.arange(1,2,0.2)
t_range = np.append(t_range,np.arange(2,2.6,0.1))
t_range = np.append(t_range,np.arange(2.6,3.5,0.2))
```

```python
m10_zakres = np.load(r'C:\\Users\\mazur\\OneDrive\\Dokumenty\\GitHub\\Fizyka-ukladow-zlozonych\\danem\\m10_zakres.npy', allow_pickle=True)
m10_1=[]
for i in m10_zakres:
    m10_1.append(sum(abs(k) for k in i[10**4:])/len(i[10**4:]))
```

```python
m50_zakres = np.load(r'C:\\Users\\mazur\\OneDrive\\Dokumenty\\GitHub\\Fizyka-ukladow-zlozonych\\danem\\m50_zakres.npy', allow_pickle=True)
m50_1=[]
for i in m50_zakres:
    m50_1.append(sum(abs(k) for k in i[10**4:])/len(i[10**4:]))
```

```python
m100_zakres = np.load(r'C:\\Users\\mazur\\OneDrive\\Dokumenty\\GitHub\\Fizyka-ukladow-zlozonych\\danem\\m100_zakres.npy', allow_pickle=True)
m100_1=[]
for i in m100_zakres:
    m100_1.append(sum(abs(k) for k in i[10**4:])/len(i[10**4:]))
```

```python
plt.figure(figsize=(8,5))
plt.plot(t_range,m10_1, '-o', color='hotpink', label='L=10')
plt.plot(t_range,m50_1, '-o', color='purple', label='L=50')
plt.plot(t_range,m100_1, '-o', color='skyblue', label='L=100')
plt.grid()
plt.legend()
plt.title('Magnetyzacja jako funkcja temperatury',fontsize=15)
plt.xlabel('T', fontsize=15)
plt.ylabel('m', fontsize=15)
plt.show()
```

```python
t_range = np.arange(1,2,0.2)
t_range = np.append(t_range,np.arange(2,2.6,0.1))
t_range = np.append(t_range,np.arange(2.6,3.5,0.2))
```

```python
m10_2=[]
for i in m10_zakres:
    m10_2.append(sum(k**2 for k in i[10**4:])/len(i[10**4:]))
pod_10 = []
for i in range(len(m10_2)):
    pod_10.append((m10_2[i]-(m10_1[i])**2)*10**2/t_range[i])
```

```python
m50_2=[]
for i in m50_zakres:
    m50_2.append(sum(k**2 for k in i[10**4:])/len(i[10**4:]))
pod_50 = []
for i in range(len(m50_2)):
    pod_50.append((m50_2[i]-(m50_1[i])**2)*50**2/t_range[i])
```

```python
m100_2=[]
for i in m100_zakres:
    m100_2.append(sum(k**2 for k in i[10**4:])/len(i[10**4:]))
pod_100 = []
for i in range(len(m100_2)):
    pod_100.append((m100_2[i]-(m100_1[i])**2)*100**2/t_range[i])
```

```python
plt.figure(figsize=(8,5))
plt.plot(t_range,pod_10, '-o', color='hotpink', label='L=10')
plt.plot(t_range,pod_50, '-o', color='purple', label='L=50')
plt.plot(t_range,pod_100, '-o', color='skyblue', label='L=100')
plt.grid()
plt.legend()
plt.title('PodatnoÅ›Ä‡ magnetyczna jako funkcja temperatury',fontsize=15)
plt.xlabel('T', fontsize=15)
plt.ylabel('gamma', fontsize=15)
plt.show()
```
