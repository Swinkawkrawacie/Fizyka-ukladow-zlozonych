```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('N100dx0.02L1000.txt') as f:
    N100_data = f.readlines()
with open('N200dx0.02L1000.txt') as f:
    N200_data = f.readlines()
with open('N300dx0.02L1000.txt') as f:
    N300_data = f.readlines()
with open('N1000dx0.02L1000.txt') as f:
    N1000_data = f.readlines()
```

```python
x = []
time_100 = []
prob_100 = []
for i in N100_data:
    new = i[:-1].split('  ')
    x.append(float(new[0]))
    prob_100.append(float(new[1]))
    time_100.append(float(new[2]))
```

```python
time_200 = []
prob_200 = []
for i in N200_data:
    new = i[:-1].split('  ')
    prob_200.append(float(new[1]))
    time_200.append(float(new[2]))
```

```python
time_300 = []
prob_300 = []
for i in N300_data:
    new = i[:-1].split('  ')
    prob_300.append(float(new[1]))
    time_300.append(float(new[2]))
```

```python
time_1000 = []
prob_1000 = []
for i in N1000_data:
    new = i[:-1].split('  ')
    prob_1000.append(float(new[1]))
    time_1000.append(float(new[2]))
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, prob_100, 'o', color='hotpink', label=r'L=${10}^2$')
plt.plot(x, prob_200, 'o', color='purple', label=r'L=$2*{10}^2$')
plt.plot(x, prob_300, 'o', color='skyblue', label=r'L=$3*{10}^2$')
plt.plot(x, prob_1000, 'o', color='lime', label=r'L=${10}^3$')
plt.grid()
plt.legend()
plt.title('P(x)',fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('P(x)', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, time_100, 'o', color='hotpink', label=r'L=${10}^2$')
plt.plot(x, time_200, 'o', color='purple', label=r'L=$2*{10}^2$')
plt.plot(x, time_300, 'o', color='skyblue', label=r'L=$3*{10}^2$')
plt.plot(x, time_1000, 'o', color='lime', label=r'L=${10}^3$')
plt.grid()
plt.legend()
plt.title('\u03c4(x)',fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('\u03c4(x)', fontsize=15)
plt.show()
```
