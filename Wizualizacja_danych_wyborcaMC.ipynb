```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('N100dx0.02L1000.txt') as f:
    N100_data = f.readlines()
#with open('N1000dx0.02L1000.txt') as f:
#    N1000_data = f.readlines()
```

```python
x = []
time_100 = []
prob_100 = []
for i in N100_data:
    new = i[:-1].split('  ')
    x.append(float(new[0]))
    prob_100.append(float(new[1]))
    time_100.append(float(new[2]))
```

```python
time_1000 = []
prob_1000 = []
for i in N1000_data:
    new = i[:-1].split('  ')
    x.append(float(new[0]))
    prob_1000.append(float(new[1]))
    time_1000.append(float(new[2]))
```

```python
time_10000 = []
prob_10000 = []
for i in N10000_data:
    new = i[:-1].split('  ')
    x.append(float(new[0]))
    prob_10000.append(float(new[1]))
    time_10000.append(float(new[2]))
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, prob_100, '-o', color='hotpink', label=r'L=${10}^2$')
#plt.plot(x, prob_1000, '-o', color='purple', label=r'L=${10}^3$')
#plt.plot(x, prob_10000, '-o', color='skyblue', label=r'L=${10}^4$')
plt.grid()
plt.legend()
plt.title('P(x)',fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('P(x)', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, time_100, '-o', color='hotpink', label=r'L=${10}^2$')
#plt.plot(x, time_1000, '-o', color='purple', label=r'L=${10}^3$')
#plt.plot(x, time_10000, '-o', color='skyblue', label=r'L=${10}^4$')
plt.grid()
plt.legend()
plt.title('\u03c4(x)',fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('\u03c4(x)', fontsize=15)
plt.show()
```
